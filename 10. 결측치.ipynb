## 데이터 제외하기 dropna

```python
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
df.dropna(inplace = True) #전체 데이터 중nan을 포함하는 데이터를 삭제
```

```python
df
```

```python
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

- axis : index or columns (row를 지울지 col를 지울지 선택)
- how: any or all (default = any)

```python
df.dropna(axis='index', how='any') # nan데이터가 하나라도 있는 데이터 삭제
```

```python
df.dropna(axis='columns')
```

```python
df['학교'] = np.nan
df
```

```python
df.dropna(axis = 'columns', how='all') #데이터 전체가 nan인 경우에 삭제
```
