# 12. 데이터 수정

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

## column 수정

```python
df['학교'].replace({'북산고' : '상북고', '능남고' : '무슨고'}) #dict 형태
```

```python
df
```

```python
df['학교'].replace({'북산고' : '상북고'}, inplace=True)
df
```

```python
df['SW특기'].str.lower()
```

```python
df['SW특기'] = df['SW특기'].str.lower()
df
```

```python
df['SW특기'] = df['SW특기'].str.upper()
df
```

```python
df['학교'] = df['학교'] + '등학교' #학교데이터 + '등학교'
df
```

## column 추가

```python
df['총합'] = df['국어']+df['영어']+df['수학']+df['과학']+df['사회']
df
```

```python
90+85+100+95+85
```

```python
df['결과'] = 'Fail' #결과 컬럼을 추가하고 전체 데이터는 fail
df
```

```python
df.loc[df['총합'] > 400, '결과'] = 'Pass' # 총합이 400보다 큰 데이터ㅔ 대해서 결과를 패스로 업데이터
df
```

## column 삭제

```python
df.drop(columns=['총합']) # 총합 컬럼 삭제
```

```python
df.drop(columns=['국어', '영어', '수학'], inplace=True) #국어, 영어, 수학 컬럼을 삭제
```

```python
df
```

## cell 수정

```python
df.loc['4번', 'SW특기'] = 'Python' #4번 학생의 sw특기 데이터를 python으로 변경
df
```

```python
df.loc['5번', ['학교', 'SW특기']] = ['능남고등학교', 'C']
# 5번학생의 학교는 능남고로 SW특기는 C로 변경
df
```

## column 순서 변경

```python
cols = list(df.columns)
cols
```

```python
df = df[['결과', '이름', '학교']]
df
```

```python
df = df[[cols[-1]] + cols[0:-1]] # 0:-1은 결과 직전까지 가져오기
df
#맨 뒤에 있는 결과 컬럼을 앞으로, 나머지 컬럼을 합쳐서 순서 변경
```
